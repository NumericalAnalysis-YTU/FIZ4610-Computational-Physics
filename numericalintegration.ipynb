```python
def simp(a, b, h, f):
    F = 0
    xt = a + h
    while xt <= b:
        F = F + h/3 * (f(xt) + 4*f(xt - h) + f(xt - 2*h))
        xt = xt + 2*h
    return F

def trap(a, b, h, f):
    F = 0
    xi = a + h
    while xi <= b:
        F = F + 1 / 2 * h * (    f(xi) + f(xi - h)    )
        xi = xi + h
    return F

f=lambda x: x**5

a=3
b=10
n=1
h=(b-a)/n
trapazoidal1= trap(a,b,h,f)
simpson1=simp(a,b,h,f)
print (trapazoidal1, simpson1)
```
