# Polynomials

A polynomial is expressed as:

$p(x) = p_1 x^n + p_2 x^{n-1} + ⋯ + p_n + p_{n+1}$

where $p_1, p_2, p_3 ...$ are the coefficients of the polynomial.

We have *Linear Regression* and *Polynomial Regression*


*Linear Regression* is a special case of  Polynomial Regression

• Linear Regression is a 1.order Polynomial  (n = 1)

```python
import numpy.polynomial.polynomial as poly
p = [5.6, 8, 3.2, 0, -5.45]
r = poly.polyroots(p)
print(r)
```

```python
# data
import math
from matplotlib import pyplot as plt
# import matplotlib.pyplot as plt

x = [24, 32, 40, 48, 64, 80]
y = [4.8, 6.0, 10.2, 12.0, 18.0, 27.0]

n = len(x)

#  % compute the coefficients

sum1=0.  #x
sum2=0.  #x*x
sum3=0.  #x*y
sum4=0.  #y
sum5=0. #(y-a*x-b)**2
sum6=0.  #(y-ym)**2
for i in range (n):
  sum1 = sum1+x[i]
  sum2 += x[i]*x[i]
  sum3 += x[i]*y[i]
  sum4 += y[i]
 
  factor = n*sum2-sum1*sum1

  a = (n*sum3-sum1*sum4)/factor
  b = (sum2*sum4-sum1*sum3)/factor

  sum5 += (y[i]- a*x[i]-b)**2
  ym = sum5/n
  sum6 += (y[i]-ym)**2
#  % compute the goodness of the fit
  c=a*x[i]+b
  print(c)
  
  S = sum5
  St = sum6
  r2 = (St-S)/St
 #   print out the results

print("Fit Results\n")
print("a = \n",a)
print("b = \n",b)
print("r2= \n",r2)
plt.plot(x,y,"o--",c,"r-")
```

#Polynomial Regression


We will use the Python to find and compare the models using different orders of the polynomial.



We have only 6 data points, so a model with order higher  than 5 will make no sense.

Typically we have much more data, but this is just an 
example to demonstrate the principle of curve fitting

We want to find models on the form:


$p(x) = p_1 x^n + p_2 x^{n-1} + ⋯ + p_n x + p_{n+1}$

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [15, 10, 9, 6, 2, 0]

def model(x, a, b, c):
  y = a * x ** 2 + b * x + c
  return y 

popt, pcov = curve_fit(model, x, y)

print(popt)
print("y(x)=",popt[0], "*x^2 +",popt[1],"x + ", popt[2])
plt.plot(x,y, 'ok')

xstart = -1
xstop = 6
increment = 0.1

xmodel = np.arange(xstart,xstop,increment)

a = popt[0]
b = popt[1]
c = popt[2]

ymodel = model(xmodel, a, b, c)

plt.plot(xmodel,ymodel, 'b')
```

# Curve Fitting
Assume we want to fit some given data to 
the following model:

$y(x) = a * sin(x + b)$

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
start = 0
stop = 2*np.pi
increment = 0.5

x = np.arange(start,stop,increment)

a = 2
b = 10
np.random.seed()

y_noise = 0.2 * np.random.normal(size=x.size)

y = a * np.sin(x + b)
y = y + y_noise
plt.plot(x,y, 'or')

def model(x, a, b):
  y = a * np.sin(x + b)
  return y 

popt, pcov = curve_fit(model, x, y)
increment = 0.1

xmodeldata = np.arange(start,stop,increment)
ymodel = model(xmodeldata, *popt)
plt.plot(xmodeldata,ymodel)
```
